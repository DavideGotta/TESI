```python
def LOR(score,scores_coding,scores_int):
    num = sum(1 for score_int in scores_int if score_int > score) / len(scores_int)
    den = sum(1 for score_coding in scores_coding if score_coding > score) / len(scores_coding)
    if den == 0:
        return float("inf"), num, den
    else:
        return log(num / den), num, den
import pandas as pd
dfcoding=pd.read_csv('/home/davide/Documents/provastatsnew.txt',sep='\t',skiprows=1,header=None)


scorecoding=dfcoding[2].to_numpy()
dfint=pd.read_csv('/home/davide/Desktop/operoniscorenew.csv',skiprows=1,header=None)
scoresint=dfint[4].to_numpy()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from math import log

scores = np.arange(, max(scoresint), 0.1)
lor_values = [LOR(score, scorecoding, scoresint)[0] for score in scores]
prob_int = [LOR(score, scorecoding, scoresint)[1] for score in scores]
prob_coding = [LOR(score, scorecoding, scoresint)[2] for score in scores]

plt.figure(figsize=(10, 5))
plt.plot(scores, lor_values, label='LOR')
plt.plot(scores, prob_int, label='Prob Intergenic')
plt.plot(scores, prob_coding, label='Prob Coding')
plt.xlabel('Score')
plt.ylabel('Values')
plt.xticks(np.arange(-6, 20, 1))
plt.grid(True)
plt.legend()
plt.show()
```
