```python
def LOR(score,scores_coding,scores_int):
    num = sum(1 for score_int in scores_int if score_int>score)/len(scores_int)
    den = sum(1 for score_coding in scores_coding if score_coding>score) / len(scores_coding)
    if den == 0:
        return float("inf")
    else:
        return log(num / den)

# Open the file
with open('scorescoding.txt', 'r') as f:
    lines = f.readlines()

# Get the columns as lists
column1, column2 = zip(*(line.strip().split('\t') for line in lines))

# Convert the tuples to lists
scorecoding = list(map(float, column1))
scorerefinedcoding = list(map(float, column2))
```

```python
import pandas as pd
operoniscore=pd.read_csv('risultatiscoreoperonelexa.txt',sep='\t',header=None,skiprows=1)
operoniscore.columns=["sequence_name","operone","score","score_refined","motivo","posizione","info_geni"]
operoniscore=operoniscore[["score","score_refined"]]
scoreintergenic=operoniscore["score"].tolist()
scoreintergenic=list(map(float,scoreintergenic))
scorerefinedintergenic=operoniscore["score_refined"].tolist()
scorerefinedintergenic=list(map(float,scorerefinedintergenic))
```

```python
print(min(scorecoding),max(scorecoding))
print(min(scoreintergenic),max(scoreintergenic))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from math import log

scores = np.arange(min(scoreintergenic), max(scoreintergenic), 0.01)
lor_values = [LOR(score,scorecoding,scoreintergenic) for score in scores]

plt.figure(figsize=(25, 10))
plt.plot(scores, lor_values)
plt.xlabel('Score')
plt.ylabel('LOR(Score)')
plt.title('Plot di LOR(Score) contro lo score')
plt.xticks(np.arange(min(scores), max(scores)+1, 1))

plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from math import log

scores = np.arange(min(scorerefinedintergenic), max(scorerefinedintergenic), 0.1)
lor_values = [LOR(score,scorerefinedcoding,scorerefinedintergenic) for score in scores]

plt.figure(figsize=(20,10))
plt.plot(scores, lor_values)
plt.xlabel('Score')
plt.ylabel('LOR(Score)')
plt.title('Plot di LOR(score_refined) contro lo score_refined')
plt.xticks(np.arange(min(scores), max(scores)+1, 1))

plt.grid(True)

plt.show()
```
