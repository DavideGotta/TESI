```python
import pandas as pd
prokka="/home/davide/Downloads/prokka.tbl"
refseq="/home/davide/Downloads/NZ_CP083761.tbl"
with open(prokka) as f:
    prok = f.readlines()
with open(refseq) as f:
    ref = f.readlines()

refplus = []
refminus = []
prokplus = []
prokminus = []
for line in ref:
    if line[0].isdigit():
        if "CDS" in line:
            start, end= map(int,line.split('\t')[:2])
            if start < end:
                refplus.append((start,end))
            else:
                refminus.append((end,start))
cds_lines = {}
key = None
for line in prok:
    if line[0].isdigit():
        if "CDS" in line:
            start, end= map(int,line.split('\t')[:2])
            key= (start,end)
            if start < end:
                prokplus.append((start,end))
            else:
                prokminus.append((start,end))
    elif key:
        if key in cds_lines:
            cds_lines[key].append(line)
        else:
            cds_lines[key]=[line]
        #add all lines after CDS until ne wline start with digit
```

```python
for i,line in enumerate(ref.copy()):
    if line[0].isdigit():
        if "CDS" in line:
            start, end= map(int,line.split('\t')[:2])
            if start < end :
                for (a,b) in prokplus:
                    if  (start!=a and end==b): 
                        j=1
                        while "note" not in ref[i+j]:
                            j+=1
                        ref[i+j]=ref[i+j].replace("\n",f"  Prokka: inizio a {a}\n")
            else:
                for (a,b) in prokminus:
                    if (start!=a and end==b): 
                        j=1
                        while "note" not in ref[i+j]:
                            j+=1
                        print(ref[i+j])
                        ref[i+j]=ref[i+j].replace("\n",f"  Prokka: inizio a {a}\n")      
    
```

```python
count=0
count_genes=0
count_cds=0
for i,line in enumerate(ref.copy()):
    if line[0].isdigit():
        if "CDS" in line:
            count_cds+=1
            start, end= map(int,line.split('\t')[:2])
            if start < end :
                for (a,b) in prokplus:
                    if  (start==a and end==b) or (start!=a and end==b): 
                        break
            else:
                for (a,b) in prokminus:
                    if (start==b and end==b) or (start!=a and end==b): 
                        break
            if ref[i-2][0].isdigit() or "gene" not in ref[i-2] :
                count+=1
                print(f"ref {start} {end} prok {a} {b}")
                for l in cds_lines[(a,b)]:
                    if "gene" in l:
                        gene_name=l.split('\t')[-1][:-1]
                        print(gene_name)
                        count_genes+=1
                        j=1
                        while "note" not in ref[i+j]:
                            j+=1
                        if "Prokka" not in ref[i+j]:
                            ref[i+j]=ref[i+j].replace("\n",f"  Prokka: gene={gene_name}\n")
                        else:
                            ref[i+j]=ref[i+j].replace("\n",f",  gene={gene_name}\n")
print(count,count_cds,count_genes)
```
