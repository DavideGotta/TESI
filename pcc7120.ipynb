```python
file="/home/davide/Desktop/genomiChro/blastPCC7120.txt"
import pandas as pd
df=pd.read_csv(file,sep="\t",header=None)
df.columns=["query","subject","identity","length","mismatch","gapopen","qstart","qend","sstart","send","evalue","bitscore"]
df.drop(["mismatch","gapopen","qstart","qend","sstart","send","bitscore"],axis=1,inplace=True)
#create a new column with the old_locus_tag
df.loc[:,"old_locus_tag"]=df.loc[:,"subject"].str.extract(r"\((.*?)\)",expand=False)
#keep only the first hit for each query
#df.drop_duplicates(subset="query",keep="first",inplace=True)
#order the dataframe by evalue
df.sort_values(by="evalue",inplace=True)
#drop duplicates of old_locus_tag keeping the first with the lowest evalue
df.drop_duplicates(subset="old_locus_tag",keep="first",inplace=True)
df.set_index("old_locus_tag",inplace=True)
df["info_blastp"]='id:'+df["identity"].astype(str)+"%, len:"+df["length"].astype(str)+", evalue:"+df["evalue"].astype(str)
df.drop(columns=["subject","identity","length","evalue"],inplace=True)
df_blastp=df
df_blastp
```

```python
import pandas as pd
file="/home/davide/Downloads/provageni.txt"
#open the file and read the lines
with open(file) as f:
    lines=f.read()
lines=lines.split(" ")
#remove
lines=[x for x in lines if x!='Upregulated' and x!='Downregulated' and x!='genes']
lines=[lines[i:i+8]for i in range(0, len(lines), 8)]
df_anabena=pd.DataFrame(lines,columns=["index","gene_omologo","old_locus_tag","motivo_omologo","spacer","upstream","-10region","-10upstream"])
df_anabena.drop(columns=["index"],inplace=True)
#delete () from old_locus_tag
df_anabena["old_locus_tag"]=df_anabena["old_locus_tag"].str.replace("(","")
df_anabena["old_locus_tag"]=df_anabena["old_locus_tag"].str.replace(")","")
df_anabena.set_index("old_locus_tag",inplace=True)
df_anabena.drop(columns=["spacer","upstream","-10region","-10upstream"],inplace=True)
df_anabena["Evidenze"]="Anabaena PCC 7120: sovraespressione+EMSA"
df_anabena["Pubblicazione"]="Kumar et al., 2018. https://doi.org/10.1016/j.bbagrm.2018.07.007"
df_anabena
text=testo2.split()
locus_tags=[i for i in text if (i.startswith("al") or i.startswith("as")) and not i.startswith("allevia")]
motivi=[text[i+2] for i, j in enumerate(text) if j in locus_tags]
#given two lists as columns, create a dataframe
df2=pd.DataFrame(list(zip(locus_tags,motivi)),columns=["old_locus_tag","motivo_omologo"])
df2["Evidenze"]="Anabaena PCC 7120: Bioinformatica"
df2["Pubblicazione"]="Kumar et al., 2018. https://doi.org/10.1016/j.bbagrm.2018.07.007"
df2.set_index("old_locus_tag",inplace=True)
df_anabena=pd.concat([df_anabena,df2])
df_anabena
```
